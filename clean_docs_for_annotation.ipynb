```python
from dotenv import load_dotenv
import os
import re

load_dotenv()

DOCS_FILEPATH = os.environ['PATH_TO_DOCS'] # only looking at test docs at the moment
CLEAN_DOCS_FILEPATH = os.environ['PATH_TO_CLEAN_DOCS']

# list of keywords to filter docs
keywords = ["asthma"]

def remove_images(markdown_text):

    # remove images and replace with empty string
    images_removed = re.sub(r'\[([^\]]+)\]\(([^)]+)\)\]\(\/\)', '___IMAGE_REMOVED___', markdown_text, flags=re.MULTILINE)
    # remove images: ![image_name.png]()
    images_removed = re.sub(r'\!\[([^\]]*)\]\(([^)]*)\)', '___IMAGE_REMOVED___', images_removed, flags=re.MULTILINE)

    return images_removed


def remove_links(text):

    # remove inline links: [link text](URL)
    links_removed = re.sub(r'\[([^\]]+)\]\(([^)]+)\)', '___LINK_REMOVED___', text)
    # remove reference-style links: [link text][reference]
    links_removed = re.sub(r'\[([^\]]+)\]\[([^\]]+)\]', '___LINK_REMOVED___', links_removed)
    # remove reference definitions: [reference]: URL
    links_removed = re.sub(r'^\[[^\]]+\]: .+$', '___LINK_REMOVED___', links_removed, flags=re.MULTILINE)

    return links_removed


def remove_menus(text):

    menus_removed = text

    return menus_removed

# remove images

# convert headers and subheaders to plain text
# convert lists (maybe remove)



# create dir for cleaned docs
if not os.path.exists(CLEAN_DOCS_FILEPATH):
    os.makedirs(CLEAN_DOCS_FILEPATH)

for doc in os.listdir(DOCS_FILEPATH):

    # read file
    with open(os.path.join(DOCS_FILEPATH, doc), 'r', encoding='utf-8') as file:
        content = file.read()

    # remove images
    cleaned_doc = remove_images(content)
    # remove URL links
    cleaned_doc = remove_links(cleaned_doc)

    # remove any docs that do not have keywords in (currently only "asthma" is a keyword)

    # save cleaned doc if a keyword appears in tokenized word list
    doc_tokens = list(set(cleaned_doc.split()))
    for k in keywords:
        if k in doc_tokens:
            print(f'Keyword found: {k}')
            with open(os.path.join(CLEAN_DOCS_FILEPATH, doc), 'w', encoding='utf-8') as file:
                file.write(cleaned_doc)
            break
        else:
            print("No keywords in doc")

    # with open(os.path.join(CLEAN_DOCS_FILEPATH, doc), 'w', encoding='utf-8') as file:
    #     file.write(cleaned_doc)
```
